## Импорт библиотек

Импортируем необходимые библиотеки для работы с данными: pandas для обработки данных, numpy для численных операций, matplotlib и seaborn для визуализации. Также устанавливаем опцию отображения всех столбцов в pandas.

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)
```

## Загрузка сырых данных

Загружаем исходный датасет с объявлениями автомобилей из CSV-файла.

```python
df = pd.read_csv("../data/raw/all_car_adverts.csv")
```

## Обработка столбца year

Извлекаем год из строки, конвертируем в числовой тип, удаляем строки с пропусками в year и фильтруем автомобили с 1980 по 2024 год.

```python
df["year"] = df["year"].str.extract(r"(\d{4})")
df["year"] = pd.to_numeric(df["year"])

df = df.dropna(subset=["year"])

df["year"] = df["year"].astype(int)

df = df[(df["year"] >= 1980) & (df["year"] <= 2024)]
```

## Создание признака car_age

Вычисляем возраст автомобиля как разницу между текущим годом (2026) и годом выпуска.

```python
current_year = 2026

df["car_age"] = current_year - df["year"]
```

## Конвертация miles в числовой тип

Преобразуем столбец miles в числовой тип, игнорируя ошибки конвертации.

```python
df["miles"] = pd.to_numeric(df["miles"], errors="coerce")
```

## Удаление строк с пропусками в miles

Удаляем строки, где miles содержит пропуски, так как пробег является важным признаком.

```python
df = df.dropna(subset=["miles"])
```

## Обработка car_price

Конвертируем car_price в числовой тип и удаляем строки с пропусками в цене, так как это целевая переменная.

```python
df["car_price"] = pd.to_numeric(df["car_price"], errors="coerce")

df = df.dropna(subset=["car_price"])
```

## Удаление ненужных столбцов

Удаляем столбцы, которые не несут полезной информации для моделирования: заголовки, подзаголовки, спецификации и т.д.

```python
cols_to_drop = [
    "car_title",
    "car_sub_title",
    "car_attention_grabber",
    "car_specs",
    'car_seller',
    'car_seller_location',
    'reg',
    'car_badges',
    'variant',
    'engine_size_unit',
    'model'

]

df = df.drop(columns=cols_to_drop, errors="ignore")
```

## Определение категориальных столбцов

Получаем список столбцов с типом object (категориальные переменные) для дальнейшего преобразования.

```python
df.select_dtypes(include="object").columns
```

## One-hot encoding категориальных переменных

Преобразуем все категориальные переменные в бинарные dummy-переменные с помощью get_dummies, удаляя первый столбец для избежания мультиколлинеарности.

```python
df = pd.get_dummies(
    df,
    columns=df.select_dtypes(include="object").columns.to_list(),
    drop_first=True
)
```

## Просмотр обработанных данных

Выводим первые строки датасета после преобразований для проверки корректности обработки.

```python
df.head()
```

## Заполнение пропусков в engine_vol

Заполняем пропуски в объеме двигателя средним значением по марке автомобиля.

```python
df["engine_vol"] = df.groupby([col for col in df.columns if col.startswith("make_")]).engine_vol.transform(lambda x: x.fillna(x.mean()))
```

## Заполнение пропусков в engine_size

Заполняем пропуски в размере двигателя средним значением по марке автомобиля.

```python

df["engine_size"] = df.groupby([col for col in df.columns if col.startswith("make_")]).engine_size.transform(lambda x: x.fillna(x.mean()))
```

## Заполнение пропусков в num_owner

Заполняем пропуски в количестве владельцев медианным значением.

```python
df["num_owner"].fillna(df["num_owner"].median(), inplace=True)
```

## Информация о датасете после обработки

Выводим общую информацию о датасете: количество строк, столбцов, типы данных.

```python
df.info()
```

## Поиск оставшихся пропусков

Определяем столбцы, которые все еще содержат пропуски.

```python
missing_cols = df.columns[df.isnull().any()]
print(missing_cols)
```

## Заполнение пропусков в car_seller_rating

Заполняем пропуски в рейтинге продавца 25-м перцентилем.

```python
percentile_25 = df["car_seller_rating"].quantile(0.25)
df["car_seller_rating"].fillna(percentile_25, inplace=True)
```

## Проверка количества пропусков

Выводим количество пропусков по каждому столбцу.

```python
df.isnull().sum()
```

## Финальная обработка пропусков

Автоматически заполняем оставшиеся пропуски: для категориальных переменных - модой, для числовых - медианой.

```python
# Проверяем наличие пропусков
missing_cols = df.columns[df.isnull().any()]

for col in missing_cols:
    if df[col].dtype == "bool" or df[col].dtype == "object":
        # Категориальные: заменяем на самый популярный (мода)
        mode = df[col].mode()[0]
        df[col].fillna(mode, inplace=True)
    else:
        # Числовые: заменяем на медиану
        median = df[col].median()
        df[col].fillna(median, inplace=True)

# Проверяем, остались ли пропуски
df.isnull().sum()
```

## Сохранение обработанного датасета

Сохраняем очищенный и подготовленный датасет в CSV-файл для дальнейшего использования в моделировании.

```python
df.to_csv("../data/processed/cars_processed.csv", index=False)
```

## Импорт RandomForest для анализа важности признаков

Импортируем RandomForestRegressor для оценки важности признаков.

```python
from sklearn.ensemble import RandomForestRegressor
```

## Разделение на признаки и целевую переменную

Разделяем датасет на матрицу признаков X и вектор целевой переменной y (цена автомобиля).

```python
X = df.drop("car_price", axis=1)
y = df["car_price"]
```

## Обучение RandomForest для оценки важности

Обучаем модель RandomForest на всем датасете для анализа важности признаков.

```python
rf = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)

rf.fit(X, y)
```

## Создание DataFrame с важностью признаков

Создаем таблицу с названиями признаков и их важностью по версии RandomForest.

```python
feature_importance = pd.DataFrame({
    "feature": X.columns,
    "importance": rf.feature_importances_
})
```

## Сортировка признаков по важности

Сортируем признаки по убыванию важности для анализа наиболее значимых факторов.

```python
feature_importance = feature_importance.sort_values(
    "importance",
    ascending=False
)

feature_importance
```

## Выделение малозначимых признаков

Отбираем признаки с важностью менее 1% для возможного удаления.

```python
low_importance = feature_importance[
    feature_importance["importance"] < 0.01
]

low_importance
```

## Выводы по feature engineering

В результате обработки данных для предсказания цен автомобилей были выполнены следующие шаги:

1. **Очистка данных**: Удалены нерелевантные столбцы, обработаны пропуски в ключевых переменных (год, пробег, цена).

2. **Создание новых признаков**: Добавлен возраст автомобиля (car_age) как более информативный признак по сравнению с годом выпуска.

3. **Кодирование категориальных переменных**: Применен one-hot encoding для преобразования категориальных признаков в бинарные.

4. **Обработка пропусков**: Заполнены пропуски в числовых переменных средними/медианными значениями по группам (марка автомобиля), в категориальных - модой.

5. **Анализ важности признаков**: С помощью RandomForest определены наиболее значимые признаки для предсказания цены.

**Ключевые выводы:**
- Датасет очищен от пропусков и готов к моделированию
- Создано большое количество бинарных признаков из категориальных переменных
- Выявлены признаки с низкой важностью (<1%), которые можно рассмотреть для удаления
- Наиболее важные признаки (вероятно, пробег, возраст, объем двигателя) оказывают наибольшее влияние на цену

Обработанный датасет сохранен в `cars_processed.csv` и готов для этапа моделирования.
